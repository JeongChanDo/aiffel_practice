```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import dlib
```

# read image and show

```python
my_img_path = os.getenv("HOME")+"/aiffel/camera_sticker/images/image.jpg"
img_bgr = cv2.imread(my_img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_show = img_rgb.copy()
plt.imshow(img_rgb)
```

# find face using hog feature with dlib

```python
detector_hog = dlib.get_frontal_face_detector()
face_img = img_show.copy()
dlib_rects = detector_hog(img_rgb, 1)
print(dlib_rects)
for dlib_rect in dlib_rects:
    left = dlib_rect.left()
    top = dlib_rect.top()
    right = dlib_rect.right() //2
    bottom = dlib_rect.bottom() //2
    cv2.rectangle(face_img, (left, top), (right, bottom), (0, 255, 0), 2)

plt.imshow(face_img)
```

# landmark estimation using dlib's pretrained model
- find 68 face landmarks for attaching srickers
- we use pretrained model based on regression tree from dlib

```python
model_path = "models/shape_predictor_68_face_landmarks.dat"
landmark_predictor = dlib.shape_predictor(model_path)
feature_img = img_show.copy()

list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))

for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(feature_img, point, 2, (0, 255, 255), -1) # yellow

plt.imshow(feature_img)
plt.show()
```

# attach cat whisker
- find nose location using landmark
- attach cat whisker to our face

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # nose center index : 30
    x = landmark[30][0]
    y = landmark[30][1] - dlib_rect.width()//2
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```
