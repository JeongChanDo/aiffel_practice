# 1. 라이브러리 및 함수 정의
## 1.1 라이브러리 임포트

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")
```

## 1.2 데이터 로드 및 전처리

```python
def load_preprocess(file_name):  
    path = os.getenv("HOME") + "/aiffel/stock_prediction/data/"+file_name
    ts = pd.read_csv(path, index_col='Date', parse_dates=True)[["Close"]]
    ts=ts.interpolate(method='time')
    ts_log = ts.copy()
    ts_log["Close"] = np.log1p(ts_log["Close"])
    return ts_log
```

## 1.3 시계열 안정성 분석

```python
def analysis_station(ts_log):
    plot_rolling_statistics(ts_log, window=30)
    print("before timeseries decomposition")
    augmented_dickey_fuller_test(ts_log)

    from statsmodels.tsa.seasonal import seasonal_decompose
    decomposition = seasonal_decompose(ts_log, model='multiplicative', period = 30) 

    trend = decomposition.trend
    seasonality = decomposition.seasonal
    residual = decomposition.resid

    plt.subplot(411)
    plt.plot(ts_log,label="origin")
    plt.subplot(412)
    plt.plot(trend, label="Trend")
    plt.subplot(413)
    plt.plot(seasonality, label="Seasonal")
    plt.subplot(414)
    plt.plot(residual, label="residual")
    plt.show()
    print("after timeseries decomposition")
    residual.dropna(inplace=True)
    augmented_dickey_fuller_test(residual)
```

## 1.4 데이터셋 분리

```python
def train_test_split(ts_log, plot=False):
    train_data, test_data = ts_log[:int(len(ts_log) * 0.9)], ts_log[int(len(ts_log) * 0.9):]
    if plot == True:
        plt.figure(figsize=(10,6))
        plt.plot(ts_log, c='r', label="training")
        plt.plot(test_data, c="b", label="test")
        plt.legend()
    return train_data, test_data
```

## 1.5 AR시차 p, 차분 d, MA시차 q 찾기

```python
def find_pdq(ts_log):
    from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

    plot_acf(ts_log)
    plot_pacf(ts_log)
    plt.show()
    diff_1 = ts_log.diff(1).iloc[1:]
    diff_1.plot(title="diff 1")
    augmented_dickey_fuller_test(diff_1)
```

## 1.6 학습 및 예측

```python
def fit_predict(train_data, test_data, p=1,d=1,q=0):
    from statsmodels.tsa.arima_model import ARIMA

    model = ARIMA(train_data, order=(p, d, q))
    fitted_model = model.fit(disp=-1)
    print(fitted_model.summary())

    fc, se, conf = fitted_model.forecast(len(test_data), alpha=0.05) 

    fc_series = pd.Series(fc, index=test_data.index)
    lower_series = pd.Series(conf[:, 0], index=test_data.index) 
    upper_series = pd.Series(conf[:, 1], index=test_data.index)

    plt.figure(figsize=(9,5), dpi=100)
    plt.plot(train_data, label='training')
    plt.plot(test_data, c='b', label='actual price')
    plt.plot(fc_series, c='r',label='predicted price')
    plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.10)
    plt.legend()
    plt.show()
    return fc
```

## 1.7 예측 성능 평가

```python
def evaluate_ts(fc):
    from sklearn.metrics import mean_squared_error, mean_absolute_error
    import math

    mse = mean_squared_error(np.exp(test_data), np.exp(fc))
    print('MSE: ', mse)

    mae = mean_absolute_error(np.exp(test_data), np.exp(fc))
    print('MAE: ', mae)

    rmse = math.sqrt(mean_squared_error(np.exp(test_data), np.exp(fc)))
    print('RMSE: ', rmse)
    mape = np.mean(np.abs(np.exp(fc.reshape(-1, 1)) - np.exp(test_data).values)/np.abs(np.exp(test_data)).values)
    print('MAPE: {:.2f}%'.format(mape*100))
```

## 1.7 안정성 체크

```python
from statsmodels.tsa.stattools import adfuller

def augmented_dickey_fuller_test(timeseries):
    # statsmodels 패키지에서 제공하는 adfuller 메소드를 호출합니다.
    dftest = adfuller(timeseries, autolag='AIC')  
    
    # adfuller 메소드가 리턴한 결과를 정리하여 출력합니다.
    print('Results of Dickey-Fuller Test:')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

## 1.8 구간 통계치 시각화

```python
def plot_rolling_statistics(timeseries, window=31.):
    
    rolmean = timeseries.rolling(window=window).mean()
    rolstd = timeseries.rolling(window=window).std()
    
    orig = plt.plot(timeseries, color='blue',label='Original')    
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
```

# 2. 삼성 주가 예측
## 2.1 데이터 로드, 전처리 + 안정성 분석
- 시계열 데이터를 분해 하기 이전에 안정적인지 확인한 결과 pvalue가 0.96으로 시계열이 안정적이지 않다는 귀무 가설을 기각하지 못함. -> 시계열 데이터가 안정적이지 못하다.
- 하지만 시계열 데이터를 분해 후 잔차(추세와 계절성을 제거)를 보았을때 pvalue가 매우 낮으므로 안정적인 시계열임을 확인

```python
file_name = "005930.KS.csv"
ts_log = load_preprocess(file_name)
analysis_station(ts_log)
```

## 2.2 ARIMA 모델 파라미터  탐색
- pcaf 그래프를 볼때 p=1이 적합
- diff 1에 대한 가설 검정 결과 pvalue가 0.05 이하로 1차 차분으로도 시계열이 안정적임을 확인

```python
train_data, test_data = train_test_split(ts_log)
# p = 1, d = 1, q = 0으로 하자
find_pdq(ts_log)
```

## 2.3 시계열 학습 및 평가 결과
- ARIMA모델로 예측한 결과 MAPE는 6.2%가 나옴

```python
fc = fit_predict(train_data,test_data, p=1,d=1,q=0)
evaluate_ts(fc)
```
