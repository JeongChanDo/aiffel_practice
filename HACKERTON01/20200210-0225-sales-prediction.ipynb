```python
# Basic data loading
path = "data/"
df_shops_en = pd.read_csv(path + 'shops_en.csv', index_col='shop_id')
df_items_en = pd.read_csv(path + 'items_en.csv', index_col='item_id')
df_itemcat_en = pd.read_csv(path + 'item_categories_en.csv', index_col='item_category_id')
```

```python
from datetime import datetime

def date_processing(df):
    df["date"] = df["date"].apply(lambda date_str : datetime.strptime(date_str, "%d.%m.%Y"))

    df["year"] = df["date"].apply(lambda date : date.year)
    df["month"] = df["date"].apply(lambda date : date.month)
    df["day"] = df["date"].apply(lambda date : date.day)
    df = df.drop(columns=["date"])
    return df
```

```python
df_train = date_processing(df_train)
```

## 1.2. 가게 정보를 추가하여 시각화해보자

```python
print(df_shops.head())
print(df_shops_en)
```

```python
df_shops_isalnum = df_shops_en["shop_name"].apply(lambda x :  str.strip(''.join(char for char in x if char.isalnum() or char == ' ')))
df_shops_isalnum.head()
```

```python
df_shops_en["city"] = df_shops_isalnum.apply(lambda x: x.split(' ')[0])
df_shops_en.head()
```

```python
city_cnt = df_shops_en.groupby("city").count()
city_cnt = city_cnt.reset_index()
city_cnt
```

```python
"""
총 도시 개수

대부분의 가게들이 모스코바에 있다.
지역에 따라 구입량에 차이가 존재할 것으로 보인다.

지역별 물품 판매량은 얼마나 될까?
"""
plt.figure(figsize=(15,6))
ax = sns.barplot(x="city", y="shop_name", data= city_cnt)
ax.tick_params(labelsize=6)
plt.show()
```

```python
df_train_join = df_train.set_index("shop_id").join(df_shops_en)
df_train_join = df_train_join.drop(columns=["shop_name"])
df_train_join
```

```python
city_item_cnt = df_train_join.groupby(["city"])["item_cnt_day"].sum()
city_item_cnt = city_item_cnt.reset_index()
city_item_cnt["item_cnt_ratio"] = 0
city_item_cnt["item_cnt_ratio"] = city_item_cnt["item_cnt_day"].apply(lambda x : x/city_item_cnt["item_cnt_day"].sum())
city_item_cnt
```

```python
"""
도시별 물품 판매량 비율


도시별 가개 개수와 유사하나 가개가 적은 도시일수록 적고, 많은 도시일수록 더 크게 보임
"""
plt.figure(figsize=(15,6))
ax = sns.barplot(x="city", y="item_cnt_ratio", data=city_item_cnt)
ax.tick_params(labelsize=6)
plt.show()
```

```python
city_date_item_cnt = df_train_join.groupby(["city","date_block_num"])["item_cnt_day"].sum()
city_date_item_cnt = city_date_item_cnt.reset_index()
city_date_item_cnt
```

```python
"""
시간 흐름에 따른 도시별 판매량(도시, 판매량)

모스코바가 가장 크게 팔리고 있으며, 대채로 물품 판매량들이 10000이하
"""
plt.figure(figsize=(15,10))
sns.lineplot(x="date_block_num", y="item_cnt_day", hue="city", data=city_date_item_cnt)
```
