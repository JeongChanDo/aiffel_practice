# 전처리
## 1. 테스트 데이터와 관련 된 shop_id, item_id 일치하는 훈련 데이터 추출

```python
# 테스트 데이터 외 존재하는 데이터를 train 데이터에서 제거합니당.
def check_unique_shop_item(df, name="train"):
    print("####",name,"shop_id unique ####")
    print(df["shop_id"].unique().shape)
    print(df["shop_id"].unique())

    print("\n####",name,"item_id unique ####")
    print(df["item_id"].unique().shape)
    print(df["item_id"].unique())
    print("-----------------\n")
```

```python
# 먼저 train data와 test data를 비교해봅시다.

print("#########train에서 test와 id가 동일한 것들만 추출하기전########")
check_unique_shop_item(df_train, name="train")
check_unique_shop_item(df_test, name="test")
```

```python
# 제거하는 작업을 시작합니다.

def extract_from_test_unique(train, test):
    # train데이터 중 test데이터에 있는 shop_id만 남깁니다.
    test_shops = test.shop_id.unique()
    train = train[train.shop_id.isin(test_shops)]
    print("shop_id 전처리 후 :",train.shape) # 2935849 -> 2413246

    # train데이터 중 test데이터에 있는 item_id만 남깁니다.
    test_items = test.item_id.unique()
    train = train[train.item_id.isin(test_items)]
    print("item_id 전처리 후 :",train.shape) # 2413246 -> 1224439
    return train, test
```

```python
# 제거 후 데이터를 확인합니다.

df_train, df_test = extract_from_test_unique(df_train, df_test)
print("#########train에서 test와 id가 동일한 것들만 추출한 후########")
check_unique_shop_item(df_train, name="train")
check_unique_shop_item(df_test, name="test")
```

## 2. 카테고리형 변수 라벨 인코딩

```python
# shop_id, item_id가 수치형으로 인식되지 않게 라벨 인코딩을 진행합니다.

from sklearn.preprocessing import LabelEncoder

def encoder_df(train, test):
    train_enc = train.copy()
    test_enc = test.copy()
    shop_enc = LabelEncoder()
    item_enc = LabelEncoder()
    shop_enc.fit(train["shop_id"])
    item_enc.fit(test["item_id"])#test의 item_id가 유니크한게 더 많으므로 // test data의 item_id가 train data보다 많았음
    train["shop_id"] = shop_enc.transform(train["shop_id"])
    test["shop_id"] = shop_enc.transform(test["shop_id"])
    train["item_id"] = item_enc.transform(train["item_id"])
    test["item_id"] = item_enc.transform(test["item_id"])
    return train, test, shop_enc, item_enc
```

```python
df_train_enc, df_test_enc, shop_enc, item_enc = encoder_df(df_train, df_test)
```

```python
print("####### shop_id  ########")
print("train shop_id decoded/encoded")
print("inverse trasform of encoded shop_id : ", shop_enc.inverse_transform(df_train["shop_id"]))
print("encoded  : ", df_train_enc["shop_id"].values)
print("\n\ntest shop_id decoded/encoded")
print("inverse trasform of encoded shop_id : ", shop_enc.inverse_transform(df_test["shop_id"]))
print("encoded  : ", df_test["shop_id"].values)
print("\n\n####### item_id  ########")
print("train item_id decoded/encoded")
print("inverse trasform of encoded item_id : ", item_enc.inverse_transform(df_train["item_id"]))
print("encoded  : ", df_train_enc["item_id"].values)
print("\n\ntest item_id decoded/encoded")
print("inverse trasform of encoded item_id : ", item_enc.inverse_transform(df_test["item_id"]))
print("encoded  : ", df_test["item_id"].values)
```

```python
from sklearn.preprocessing import LabelEncoder
from datetime import datetime

# 데이터 전처리 함수입니다.
def date_processing(df):
    df["date"] = df["date"].apply(lambda date_str : datetime.strptime(date_str, "%d.%m.%Y")) # date를 datetime으로 변환합니다.

    df["year"] = df["date"].apply(lambda date : date.year) # 연도만 따로 빼서 컬럼으로 저장합니다.
    df["month"] = df["date"].apply(lambda date : date.month) # 월만 빼서 컬럼으로 저장합니다.
    df["day"] = df["date"].apply(lambda date : date.day) # 일만 빼서 컬럼으로 저장합니다.
    df = df.drop(columns=["date"]) # data 컬럼을 삭제합니다.
    df = df[["year", "month","day", "date_block_num", "shop_id", "item_id", "item_price", "item_cnt_day"]] # 데이터프레임 순서를 조정합니다.
    return df


df_train_preprocessed = df_train_enc.copy()
df_train_preprocessed = date_processing(df_train_preprocessed)
df_train_preprocessed.head() # 전처리된 데이터프레임입니다.

print(len(df_train_preprocessed['month'].unique())) # 12개월 모두 있습니다
```

```python
#df_train_preprocessed["item_cnt_month"] = df_train_preprocessed.groupby(["year", "month", "shop_id", "item_id"])["item_cnt_day"].sum().reset_index()["item_cnt_day"]
```

```python
# 일 단위로 count되던 판매량을 월단위로 묶어줍니다.
df_train_month = df_train_preprocessed.groupby(["year", "month", "date_block_num", "shop_id", "item_id"])["item_cnt_day"].sum().reset_index()
```

```python
# 일일 판매량을 월별 판매량으로 바꿔줍니다.
df_train_month = df_train_month.rename(columns={"item_cnt_day": "item_cnt_month"})
df_train_month
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from math import ceil
import warnings
warnings.filterwarnings(action='ignore')


# 시각화를 어떻게 해야 하나
train = df_train_month.copy()

# 훈련데이터 중에서 shop_id 컬럼과 year, month를 기준으로 묶은 뒤, item_cnt_dat를 모두 더합니다. 그 뒤 인덱스를 리셋합니다.
def group(df, year=2013) :
    grouped = pd.DataFrame(df.groupby(['shop_id', 'year', 'month'])['item_cnt_month'].sum().reset_index())
    grouped = grouped.groupby(['year']).get_group(year)
    return grouped

grouped = group(train,2013)

# 2013년 그래프를 그려봅시다!! 
# shop을 기준으로 그립니다!!
fig, axes = plt.subplots(nrows=3, ncols=3, sharex=True, sharey=True, figsize=(16,20))
num_graph = 10
id_per_graph = ceil(grouped.shop_id.max() / num_graph) # 이것의 의미?? 특정 범위를 한정하는 것 같은데 나중에 자세히 보자.
count = 0 # 순서 넘어갈 때 인덱스로 사용


# 아까와 동일하게 그린다. 
# 다만 시간대는 x축은 momth로 주고, y축은 일일 판매량 그대로이다. 데이터는 가게 아이디로 한다.
for i in range(3):
    for j in range(3):
        sns.pointplot(x='month', y='item_cnt_month', hue='shop_id', 
                      data=grouped[np.logical_and(count*id_per_graph <= grouped['shop_id'], grouped['shop_id'] < (count+1)*id_per_graph)], ax=axes[i][j])
        count += 1
```

# 여기서부터는 모델학습 및 예측입니다.

```python
## 12월 빼기 // 12월은 비정상적인 판매를 보이기 때문에 제거합니다.
df_train_month = df_train_month[df_train_month.month != 12]
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
random_state = 200

month_X = df_train_month.iloc[:,:5]
month_y = df_train_month.iloc[:,5]

def month_predict(model, month_X, month_y, kind="default", print_len=7):
    X_train, X_val, y_train, y_val = train_test_split(month_X, month_y, test_size=0.2, random_state=random_state)
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    model_name = model.__class__.__name__
    print("#### " +model_name+" month prediction("+kind+") ####")
    print("y_train[:{}] : {}".format(print_len, np.round(y_train[:print_len].values,1)))
    print("y_pred_train[:{}] : {}".format(print_len, np.round(y_pred_train[:print_len],1)))
    print()
    y_pred_val = model.predict(X_val)
    print("y_val[:{}] : {}".format(print_len, np.round(y_val[:print_len].values,1)))
    print("y_pred_val[:{}] : {}".format(print_len, np.round(y_pred_val[:print_len],1)))
    rmse = np.sqrt(mean_squared_error(y_pred_val, y_val))
    print("rmse : {:.6f}".format(rmse))
    print("----------\n")
    return model
```

```python
from sklearn.ensemble import RandomForestRegressor
month_regressor = RandomForestRegressor(random_state=random_state)
month_regressor = month_predict(month_regressor, month_X, month_y)
```

```python
df_test["year"] = 2015
df_test["month"] = 11
df_test["date_block_num"] = 34

df_test = df_test[["year", "month", "date_block_num", "shop_id", "item_id"]]
df_test
```

```python
df_test["item_cnt_month"] = month_regressor.predict(df_test)
df_test
```
