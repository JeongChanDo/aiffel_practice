# read speech data

```python
import numpy as np
import os

speech_data_path = os.getenv("HOME")+"/aiffel/speech_recognition/data/speech_wav_8000.npz"
speech_data = np.load(speech_data_path)
print(speech_data)
speech_data.__dict__
```

# split speech data to X, y and get a shape of these

```python
X = speech_data["wav_vals"]
y = speech_data["label_vals"]
print(X.shape)
print(y.shape)
```

# listen one and get a label

```python
import IPython.display as ipd

idx = 2
sample_rate = X.shape[1]
print("curr idx({}) label : {}".format(idx, y[idx]))
ipd.Audio(X[idx],rate=sample_rate)
```

# make word_to_index and index_to_word

```python
target_list = ['yes', 'no', 'up', 'down', 'left', 'right', 'on', 'off', 'stop', 'go']
target_list.append("unknown")
target_list.append("silence")

word_to_index = {}
for i, label in enumerate(target_list):
    word_to_index[label] = i

index_to_word = {index: item for item, index in word_to_index.items()}
print(index_to_word)
print(word_to_index)
```

# convert y's label data to index of label

```python
print("before conversion")
print(y[:5])


print("\n\nafter conversion")
y_indexed = np.array([word_to_index[y[row][0]] for row in range(y.shape[0])])
print(y_indexed[:5])
```

# split x, y to train & test set and tf wav to spectogram

```python
from sklearn.model_selection import train_test_split
import librosa

def convert_wav2spec(X, fft_size =258):
    X_spec = []
    for wav in X:
        spec = np.abs(librosa.stft(wav, n_fft=fft_size))
        X_spec.append(spec)
    X_spec = np.array(X_spec)
    return X_spec

X_spec = convert_wav2spec(X)
input_shape = (X_spec.shape[1], X_spec.shape[2], 1)
X_train, X_test, y_train, y_test = train_test_split(X_spec, y_indexed, test_size=0.1)

X_train = X_train.reshape(-1, input_shape[0], input_shape[1], 1)
X_test = X_test.reshape(-1, input_shape[0], input_shape[1], 1)


train_sz, test_sz = X_train.shape[0], X_test.shape[0]


print("X_train shape : ", X_train.shape)
print("Y_train shape : ", y_train.shape)
print("X_test shape : ", X_test.shape)
print("y_test shape : ", y_test.shape)
print("input shape : ", input_shape)
```

# encode the labels as one hot and preparation of dataset

```python
import tensorflow as tf
from tensorflow.data import Dataset


batch_size = 256

def one_hot(x, label):
    label = tf.one_hot(label, depth=12)
    return x, label


train_dataset = Dataset.from_tensor_slices((X_train, y_train))
train_dataset = train_dataset.map(one_hot)
train_dataset = train_dataset.repeat().batch(batch_size=batch_size)
test_dataset = Dataset.from_tensor_slices((X_test, y_test))
test_dataset = test_dataset.map(one_hot)
test_dataset = test_dataset.batch(batch_size=batch_size)

del X_train, y_train, X_test, y_test, X_spec, X, y, y_indexed
```

# audio classification of specgoram using cnn

ref : https://medium.com/x8-the-ai-community/audio-classification-using-cnn-coding-example-f9cbd272269e

```python
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Activation, Conv2D, MaxPool2D, \
                                Dropout, BatchNormalization, Flatten

tf.keras.backend.clear_session()
input_tensor = layers.Input(shape=input_shape)

x = Conv2D(32, kernel_size=(3,3), padding="same", activation="relu")(input_tensor)
x = MaxPool2D(pool_size=(2, 2))(x)
x = Conv2D(64, kernel_size=(3,3), padding="same", activation="relu")(x)
x = MaxPool2D(pool_size=(2, 2))(x)
x = Dropout(0.3)(x)
x = Conv2D(128, kernel_size=(3,3), padding="same", activation="relu")(x)
x = MaxPool2D(pool_size=(2, 2))(x)
x = Dropout(0.3)(x)
x = Flatten()(x)
x = Dense(128)(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
output_tensor = Dense(len(target_list))(x)

model = tf.keras.Model(input_tensor, output_tensor)
model.summary()
```

# set loss, optimizer, metrics and define calllback fn

```python
optim = tf.keras.optimizers.Adam(1e-5)

checkpoint_dir = os.getenv("HOME") +"/aiffel/speech_recognition/models/spec"
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
             optimizer=optim,
             metrics=["accuracy"])

callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_dir,
                                             save_weights_only=True,
                                             monitor="val_loss",
                                             mode="auto",
                                             save_best_only=True,
                                             verbose=1)
```

# train datasets
- 뒷 내용의 오류를 수정하는 동안 여러번 학습을 이미 반복을 해놓아 5에폭 만 추가로 학습하였습니다.

```python
max_epochs = 5

history = model.fit(train_dataset, epochs=max_epochs,
                   steps_per_epoch=train_sz//batch_size,
                   validation_data=test_dataset,
                   validation_steps=test_sz//batch_size,
                   callbacks=[callback])

```

# plot accuracy, loss of model with history

```python
import matplotlib.pyplot as plt

accuracy = history.history["accuracy"]
validation_accuracy = history.history["val_accuracy"]
loss = history.history["loss"]
validation_loss = history.history["val_loss"]

epochs_range = range(len(accuracy))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, accuracy, label="training accuracy")
plt.plot(epochs_range, validation_accuracy, label="validation accuracy")
plt.legend(loc="lower right")
plt.title("training and validation accuracy")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label="training loss")
plt.plot(epochs_range, validation_loss, label="validation loss")
plt.legend(loc="lower right")
plt.title("training and validation loss")
plt.show()
```

# load weights and evaluate my model

```python
model.load_weights(checkpoint_dir)
results = model.evaluate(test_dataset)
print("loss val : {:.3f}".format(results[0]))
print("accuracy val : {:.4f}".format(results[1]*100))
```

# recover from spectrogram to audio and check prediction result

```python
import librosa.display
idx = np.random.choice(test_sz//batch_size, size=1)

for specs, labels in test_dataset.take(1):
    batch_specs = specs.numpy()
    batch_labels = labels.numpy()
print("batch spectogram shape : ", batch_specs.shape)
print("batch labels shape : ", batch_labels.shape)



spec = np.squeeze(batch_specs[idx])
y_pred = model(batch_specs[idx].reshape(-1, input_shape[0], input_shape[1], 1))
label = batch_labels[idx]

librosa.display.specshow(librosa.amplitude_to_db(spec,ref=np.max), x_axis='time')
plt.title("spectogram")
plt.colorbar(format="%+2.0f dB")
plt.xticks(range(0,1))
plt.tight_layout()
plt.show()

audio = librosa.istft(spec,length=8000)

print("istft shape : ", audio.shape)
print("true label : ", index_to_word[np.argmax(label)])
print("predicted label : ", index_to_word[np.argmax(y_pred)])

ipd.Audio(audio,rate=8000)


```
